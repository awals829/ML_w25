### Q1. a)

```python
u = np.array([[1],[-2]])
print(f"u = \n{u}")
```

```python
v = np.array([[2],[-5]])
print(f"v = \n{v}")
```

```python
scaler_u = 4

scaled_u = scaler_u * u
print(f"{scaler_u}u = \n{scaled_u}")
```

```python
scaler_v = 3

scaled_v = scaler_v * v
print(f"{scaler_v}v = \n{scaled_v}")
```

```python
q1a_result = scaled_u - scaled_v
print(f"{scaler_u}u - {scaler_v}v = \n{q1a_result}")
```

### Q1. b)

```python
q1b_dot_product = (u.T @ v)[0][0] # Get the result in the 2D array index

# Alternative
# q1b_dot_product = np.dot(u.T, v)

print(f"u . v = {q1b_dot_product}")
```

### Q1. c)

```python
norm_u = np.linalg.norm(u)
```

```python
norm_v = np.linalg.norm(v)
```

```python
cos_theta = q1b_dot_product / (norm_u * norm_v)
```

```python
theta = np.arccos(cos_theta)
```

```python
theta_degrees = np.degrees(theta)
```

```python
print(f"Dot product: {q1b_dot_product}")
print(f"||u||: {norm_u:.7f}")
print(f"||v||: {norm_v:.7f}")
print(f"arccos({cos_theta:.7f}): {theta:.7f}")
print(f"Angle theta (degrees): {theta_degrees:.7f}")
```

### Q2. a)

```python
matrix_A = np.array([[1, 3],
                     [2, -1]])

print(f"Matrix A = \n{matrix_A}")

vector_b = np.array([[1],
                     [0]])

print(f"Vector b = \n{vector_b}")
```

```python
# Compute Ab
Ab = matrix_A @ vector_b

print(f"Ab = \n{Ab}")
```

### Q2. b)

```python
A_inverse = np.linalg.inv(matrix_A)
print(f"A^-1 = \n{A_inverse}")
```

```python
identity_A = matrix_A @ A_inverse
identity_A.astype(int)
```

```python
# Ax = b, solve for x
x = A_inverse @ vector_b
print(f"x = \n {x}")
```
