```python
import numpy as np
```

### q1. a

```python
u = np.array([[1],[-2]])
print(f"u = {u}")
```

```python
v = np.array([[2],[-5]])
print(f"v = {v}")
```

```python
scaler_u = 4

scaled_u = scaler_u * u
print(f"{scaler_u}u = \n{scaled_u}")
```

```python
scaler_v = 3

scaled_v = scaler_v * v
print(f"{scaler_v}v = \n{scaled_v}")
```

```python
q1a_result = scaled_u - scaled_v
print(f"{scaler_u}u - {scaler_v}v = \n{q1a_result}")
```

### q1. b

```python
q1b_dot_product = (u.T @ v)[0][0] # Get the result in the array

# Alternative
# q1_dot_product = np.dot(u.T, v)

print(f"u . v = {q1b_dot_product}")
```

### q1. c

```python
norm_u = np.linalg.norm(u)
```

```python
norm_v = np.linalg.norm(v)
```

```python
cos_theta = q1b_dot_product / (norm_u * norm_v)
```

```python
theta = np.arccos(cos_theta)
```

```python
theta_degrees = np.degrees(theta)
```

```python
print(f"Dot product: {q1b_dot_product}")
print(f"||u||: {norm_u}")
print(f"||v||: {norm_v}")
print(f"cos(theta): {cos_theta}")
print(f"Angle theta (radians): {theta}")
print(f"Angle theta (degrees): {theta_degrees}")
```

### q1. d

```python
import matplotlib.pyplot as plt
```

```python
# Extract coordinates for plotting
u_coords = u.flatten()  # Convert to a 1D array for ease of access
v_coords = v.flatten()

# Plot the vectors
plt.figure(figsize=(10, 10))

# plt.quiver(origin x, origin y, endpoint x, endpoint y, ...)
plt.quiver(0, 0, u_coords[0], u_coords[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector u')
plt.quiver(0, 0, v_coords[0], v_coords[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector v')

# Set the graph plot limits, maximum x and y visual coordinates
plt.xlim(-6, 6)
plt.ylim(-6, 6)

# Add grid and labels
plt.grid()
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.title('Vector Plot and Angle Check')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Show plot
plt.show()
```
